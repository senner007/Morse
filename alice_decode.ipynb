```python
import re
class WordsFromText:
    def getWordsInFileName(self, txtfile):
        wordsarray = []
        with open(txtfile) as f:
            lines = f.readlines()
            for i in range(len(lines)):
                wordsarray += lines[i].split()
        self.result = wordsarray
        return self
    def getUniqueWords(self):
        self.result = list(set(self.result))
        return self
    def getWordsBelowLength(self, wordLength):
        self.result = [x for x in self.result if len(x) <=wordLength]
        return self
    def removeSpecialCharactersFromWords(self, searchQuery):
        self.result = [x for x in self.result if re.match(searchQuery, x)]
        return self
    def wordsToLowerCase(self):
        self.result = [x.lower() for x in self.result]
        return self
    def sortWords(self):
        self.result.sort()
        return self
result = (
    WordsFromText()
        .getWordsInFileName("alice.txt")
        .getWordsBelowLength(50)
#        .removeSpecialCharactersFromWords(r'[a-zA-Z0-9]+$')
        .removeSpecialCharactersFromWords(r'[a-zA-Z]+$')
        .wordsToLowerCase()
        # .getUniqueWords()
        # .sortWords()
        .result
)

result
```
